```python
import matplotlib.pyplot as plt
import pandas as pd

cases = ['DAC','NK_DAC','NK','Inf_renew'] # DAC, NK_DAC, NK or Inf_renew
#cases = ['Inf_renew']
studies = ['cost_ng','cost_h2','imp_h2','cap_co2']
#studies = ['cost_ng']
values = dict()
values['cost_h2'] = [0.075,0.1,0.125,0.150]
values['cost_ng'] = [50,100,150,200]
#values['cost_ng'] = [50,100,150,200]
values['imp_h2'] = [100,75,50,25]
values['cap_co2'] = [1.75,2,2.75,3.5,4]
```

```python
data = {}
capacities = {}
sheet_names = ['OFFSHORE','ZEEBRUGGE','INLAND','Energy vector','Interconnections']
for case in cases:
    for study in studies:
        data[case + ' ' + study] = dict()
        capacities[case + ' ' + study] = dict()
        if case =='Inf_renew' and study == 'cost_ng':
            values['cost_ng'] = [50,150,200]
        else:
            values['cost_ng'] = [50,100,150,200]
        for value in values[study]:
            data[case + ' ' + study][value] = dict()
            for name in sheet_names:
                #data[case + ' ' + study][value][name] = pd.DataFrame()
                data[case + ' ' + study][value][name] = pd.read_excel(case + '_' + study + '_' + str(value)+'\COE.xlsx',sheet_name=name,index_col=0)
                #data[case + ' ' + study][value][name].set_index('Unnamed: 0')
        
        
```

```python
case = 'DAC'
study = 'cost_h2'
sheet_names = ['OFFSHORE','ZEEBRUGGE','INLAND']
variables = ['Capacity','Total production']
values['cost_h2'] = [0.075,0.1,0.125,0.150]
values['cost_ng'] = [50,100,150,200]
if case =='Inf_renew' and study == 'cost_ng':
    values['cost_ng'] = [50,150,200]
values['imp_h2'] = [100,75,50,25]
values['cap_co2'] = [1.75,2,2.75,3.5,4]

evolution = {}
for name in sheet_names:
    technologies = list(data[case + ' ' + study][values[study][0]][name].index)
    evolution[name] = {}
    for technology in technologies:
        evolution[name][technology] = {}
        for value in values[study]:
            evolution[name][technology][value] = data[case + ' ' + study][value][name].loc[technology,variable]
    if name in ['OFFSHORE','ZEEBRUGGE']:
        pd.DataFrame.from_dict(evolution[name]).plot()
        plt.title(case + ' ' + study + ' ' + name)
        plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5),prop={'size': 10})
        plt.xlabel(study)
        plt.ylabel(variable + ' GW or kt/h')
        plt.show()
    else:
        nodes = ['PV']
        for node in nodes:
            plt.plot(list(evolution[name][node].keys()),list(evolution[name][node].values()),label = node)
        plt.title(case + ' ' + study + ' ' + name)
        plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5),prop={'size': 10})
        plt.xlabel(study)
        plt.ylabel(variable + ' GW or kt/h')
        plt.show()
        nodes = ['CCGT','OCGT','SMR','BIOMETHANE','NUCLEAR','ELECTROLYSIS_PLANTS']
        for node in nodes:
            plt.plot(list(evolution[name][node].keys()),list(evolution[name][node].values()),label = node)
        plt.title(case + ' ' + study + ' ' + name)
        plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5),prop={'size': 10})
        plt.xlabel(study)
        plt.ylabel(variable + ' GW or kt/h')
        plt.show()
        nodes = ['WIND_ONSHORE','DAC','PCCC_CCGT','PCCC_OCGT','PCCC_SMR']
        for node in nodes:
            plt.plot(list(evolution[name][node].keys()),list(evolution[name][node].values()),label = node)
        plt.title(case + ' ' + study + ' ' + name)
        plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5),prop={'size': 10})
        plt.xlabel(study)
        plt.ylabel(variable + ' GW or kt/h')
        plt.show()
```
